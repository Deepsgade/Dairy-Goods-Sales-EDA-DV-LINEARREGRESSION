```python
x = df['Quantity Sold (liters/kg)'].to_frame()  #independent variable
y = df['Approx. Total Revenue(INR)']            # dependent variable
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y ,color='y')
plt.xlabel('Quantity Sold (liters/kg)')
plt.ylabel('Approx. Total Revenue(INR)')
plt.show()
```

SPLIT THE DATA:TRAIN & TEST 75:25

```python
from sklearn.model_selection import train_test_split
# split the dataset into 80% for training and 20% for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=78)
```

```python
x_train.shape
```

```python
x_test.shape
```

#LinearRegression

Import sklearn

fit method

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
m = reg.coef_[0]
c = reg.intercept_
print(f"Slope / Regression co-efficient m = {m} \n\t y-intercept c = {c}")
```

```python
X = x.to_numpy()
y_pred = m*X + c
```

```python
plt.scatter(x,y, color='tomato')
plt.plot(X, y_pred, c='g')
plt.xlabel('Quantity Sold (liters/kg)')
plt.ylabel('Approx. Total Revenue(INR)')
plt.title('LinearRegression with slope')
plt.show()
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r_squared_score = r2_score(y_test, y_pred)
print("R-squared score = ", r_squared_score)
```
